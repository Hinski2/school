```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def compute_polynomial(x_values, xs):
    p = np.ones_like(x_values, dtype=np.float128) 
    for x_k in xs:
        p *= (x_values - x_k)
    return p
```

```python
x_axis_values = np.linspace(-1., 1., 500, dtype=np.float128) 

for n in range(4, 21):
    
    # wzy rwnoodlege
    equaly_spaced_nodes = np.linspace(-1., 1., n, dtype=np.float128)
    p_es = compute_polynomial(x_axis_values, equaly_spaced_nodes)
    
    # wzy czybyszewa
    k = np.arange(0, n, dtype=np.float128)
    czybyszew_nodes = np.cos((2*k + 1) * np.pi / (2*n)) # wikipedia
    p_cz = compute_polynomial(x_axis_values, czybyszew_nodes)
    
    # robienie wykresu
    plt.plot(x_axis_values, p_es, label='p_es')
    plt.plot(x_axis_values, p_cz, label='p_cz')
    plt.legend()
    plt.title(f'n = {n}')
    plt.xlabel('x')
    plt.ylabel('p(x)')
    plt.grid(True)
    plt.show()
    
    
    '''
    wnioski:
        1. efekt rungego: dla rosnocyhc wartoci n, wielomiany wykacuja coraz wiksze odchelenia na brzegach przediau [-1, 1]
        2. wzy czebyszewa s bardziej stabline, s bardziej zagszczone na kocach przedziau wic krace przedziau s bardzij stabilne 
        3. wybr wzw jest kluczowy do odpowiedniej interpolacji wielomianowej 
        
    '''
    
    
```
