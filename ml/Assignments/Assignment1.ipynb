**g) Declare `ismagic(A)` function** 
which checks if matrix $A$ is a [magic square](https://en.wikipedia.org/wiki/Magic_square) and returns a boolean. **Remark:** Don't use any for loop

```python
def ismagic(A: np.matrix):
    if len(A.shape) != 2: return 0 														# check if A is a matrix
    if A.shape[0] != A.shape[1]: return 0 												# check if A isn't a square matrix
    if not np.array_equal(sorted(A.reshape(-1)), np.arange(1, A.size + 1)): return 0 	# check if A contains numbers [1, 2, ... n] 
    if not np.all(sum(A) == sum(A[0])): return 0										# check sum in rows
    if not np.all(sum(A.T) == sum(A[0])): return 0										# chcek sum in columns
    if sum(np.diagonal(A)) != sum(A[0]): return 0										# check sum in diagonal (left top, rigth bottom)
    if sum(np.diagonal(np.fliplr(A))) != sum(A[0]): return 0							# check sum in diagonal (rigth top, left bottom)
    return 1

assert not ismagic(np.array([[1,1], [2,2]]))
assert ismagic(np.array([[2,7,6],[9,5,1],[4,3,8]]))

# some more tests to be more sure 
assert ismagic(np.array([[8, 1, 6],			# correct magic square
                         [3, 5, 7],
                         [4, 9, 2]]))

assert not ismagic(np.array([[8, 1, 1],		# incorrect magis square
                         	 [3, 5, 7],
                         	 [4, 9, 2]]))

assert ismagic(np.array([[1, 14, 4, 15],		# correct magic square
                    	 [8, 11, 5, 10], 
                         [13, 2, 16, 3],
                    	 [12, 7, 9, 6]])) 
```

The Iris data is in a so-called 'wide' format, in which each column corresponds to a variable and each row of the DataFrame corresponds to one observation. Turn it into a 'long' format in which each row is a measurement. 

Specifically, change the data layout of the IRIS dataFrame so that it has 3 columns:
- variable (one of sepal_length, sepal_width, petal_length, petal_width)
- value
- target

If you would like to learn more, [Tidy Data](https://www.jstatsoft.org/index.php/jss/article/view/v059i10/v59i10.pdf) by [Hadley Wickham](http://hadley.nz/) provides a very nice explanation of best practices for data formating.

Hint: look at reshaping functions in http://pandas.pydata.org/Pandas_Cheat_Sheet.pdf

```python
# docs: https://pandas.pydata.org/docs/reference/api/pandas.melt.html

iris_df_long = pd.melt(frame=iris_df, id_vars=['target'], value_vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], var_name='variable')
iris_df_long
```

Now create a box-plot of values that each variable takes, split by the target species.

```python
sns.catplot(iris_df_long, x="target", y="value", col="variable", kind="box")
plt.show()

sns.catplot(iris_df_long, x="target", y="value", col="variable", kind="boxen")
plt.show()

sns.catplot(iris_df_long, x="target", y="value", col="variable", kind="swarm")
plt.show()


```
