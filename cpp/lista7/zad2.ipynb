```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("zeta.csv", sep=';', names=['pos', 'imag', 'real'])
df
```

```python
pos, imag, real = df['pos'].to_numpy(), df['imag'].to_numpy(), df['real'].to_numpy()
pos, imag, real
```

```python
plt.figure(figsize=(12, 6))

plt.plot(pos, real, label="real")
plt.plot(pos, imag, label="imag")

plt.legend()
plt.axhline(0)
plt.axvline(0)

plt.show()
```
