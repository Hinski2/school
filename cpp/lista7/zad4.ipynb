```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nums.csv", sep=';', names=['unf_real', 'bin', 'norm'])
df
```

```python
unf_real, bin, norm = df['unf_real'].to_numpy(), df['bin'].to_numpy(), df['norm'].to_numpy()
unf_real, bin, norm
```

```python
plt.hist(unf_real, bins=10)
plt.title("uniform real dist")
plt.show()

plt.hist(bin, bins=10)
plt.title("binomial dist")
plt.show()

plt.hist(norm, bins=20)
plt.title("normal dist")
plt.show()
```
